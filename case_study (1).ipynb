```
def bar_chart(feature):
  satisfied=df[df['satisfaction']==1][feature].value_counts()
  not_satisfied=df[df['satisfaction']==0][feature].value_counts()
  df_1=pd.DataFrame([satisfied,not_satisfied])
  df_1.index=['satisfied','not_satisfied']
  df_1.plot(kind='bar',stacked=True,figsize=(10,5))
```

```
df=df[(df['Arrival Delay in Minutes']<500) & (df['Departure Delay in Minutes']<500)]
df
print(df.shape)
```

```
data=df.drop(['Customer ID'],axis=1)
```

```
import sklearn
from sklearn.preprocessing import LabelEncoder
data['Gender']=LabelEncoder().fit_transform(data['Gender'])
data['Customer Type']=LabelEncoder().fit_transform(data['Customer Type'])
data['Type of Travel']=LabelEncoder().fit_transform(data['Type of Travel'])
data['Class']=LabelEncoder().fit_transform(data['Class'])
data
```

```
x=data.drop(columns=['satisfaction'])
y=data["satisfaction"]
print(x.shape,y.shape)
```

```
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
x_train_mms=MinMaxScaler().fit_transform(x_train)
x_test_mms=MinMaxScaler().fit_transform(x_test)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression().fit(x_train_mms,y_train)
```

```
y_predict=LR.predict(x_test_mms)
```

```
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix
acc=accuracy_score(y_test,y_predict)
print("The accuracy is {}".format(acc))
```

```
from sklearn.model_selection import GridSearchCV
grid={"C":[0.6,0.7,0.8,0.9,1],"penalty":["l1","l2"]}
grid
```

```
score=GridSearchCV(LogisticRegression(solver = 'liblinear'),grid).fit(x_train_mms,y_train)
print(score.best_params_)
print(score.best_score_)
```

```
from sklearn.svm import SVC
linear_svm=SVC(C=1.0, kernel='linear',random_state=0)
linear_svm=linear_svm.fit(x_train_mms,y_train)
```

```
y_predict_svm=linear_svm.predict(x_test_mms)
acc_svm=accuracy_score(y_test,y_predict_svm)
print("The accuracy is {}".format(acc_svm))
```

```
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=0)
tree=tree.fit(x_train_mms,y_train)
```

```
y_predict_tree=tree.predict(x_test_mms)
acc_tree=accuracy_score(y_test,y_predict_tree)
print("The accuracy is {}".format(acc_tree))
```

```
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(criterion='gini', n_estimators=25, random_state=0)
forest=forest.fit(x_train_mms,y_train)
```

```
y_predict_forest=forest.predict(x_test_mms)
acc_forest=accuracy_score(y_test,y_predict_forest)
print("The accuracy is {}".format(acc_forest))
```

```
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb=nb.fit(x_train_mms,y_train)
```

```
y_predict_nb=nb.predict(x_test_mms)
acc_nb=accuracy_score(y_test,y_predict_nb)
print("The accuracy is {}".format(acc_nb))
```

```
from xgboost import XGBClassifier
model_xgb=XGBClassifier(gamma=5,learning_rate=0.3,max_depth=15,reg_lambda=100,n_estimators=100)
model_xgb.fit(x_train_mms,y_train)
```

```
y_predict_xgb=model_xgb.predict(x_test_mms)
acc_xgb=accuracy_score(y_test,y_predict_xgb)
print("The accuracy is {}".format(acc_xgb))
```
